```python
from datetime import datetime

from concurrent.futures import ThreadPoolExecutor
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import urllib.parse
```

```python
def get_amazon_url (lego_id:str)->str:
    query = {'k': f"lego {lego_id}"}
    query_encoded = urllib.parse.urlencode(query)
    url = f"https://www.amazon.com/s?{query_encoded}&ref=nb_sb_ss_recent_1_0_recent"
    return url


def define_title_with_lego_id (titles: list, lego_id: str):
    titles_with_lego_id = []
    for t in titles:
        if str(lego_id) in t.text:
            titles_with_lego_id.append (t)
    if len(titles_with_lego_id) == 0:
        raise ValueError ('No matches for this LEGO ID.')
    else:
        return titles_with_lego_id[0]
            
    
def scrap_price_whole (url: str, lego_id: str)->str:
    driver = webdriver.Chrome()
    driver.get(url)
    titles = driver.find_elements(By.XPATH, r'//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div/div/div/div/div/div')
    title_with_lego_id = define_title_with_lego_id (titles, lego_id)
    price_element = title_with_lego_id.find_elements(By.XPATH, r'.//*[@class="a-price"]')
    if len(price_element) == 0:
        raise ValueError ('There is no price listed for this LEGO ID.')
    else:
        price_whole = price_element[0].text
    driver.quit()
    return price_whole


def string_to_float (text: str)->float:      
    digits = '1234567890'
    for t in text:
        if t not in digits:
            text = text.replace (t,'')
    price = float (f"{text[:-2]}.{text[-2:]}")
    return price


def extract_current_amazon_price (lego_id:str)->float:
    url = get_amazon_url(lego_id)
    try:
        price_whole = scrap_price_whole (url, lego_id)
        amazon_current_price = string_to_float (price_whole)
    except ValueError:
        amazon_current_price = ''
    return amazon_current_price
            

def text_preprocessing_remove_LEGO (theme:str)->str:
    theme= str(theme).replace('LEGO® ','').replace('™','').replace('®','')
    return theme


def assessing_price_increase (catalouge_price, current_price, release_year):
    current_year = int(datetime.today().strftime('%Y'))
    avarage_yearly_price_increase = float(current_price) - catalouge_price / (current_year - release_year)
    avarage_yearly_percentage = (avarage_yearly_price_increase / float(catalouge_price))
    return avarage_yearly_percentage
```

```python
df= pd.read_csv ('data\lego_population.csv',
                index_col = 'item_number',
                skipfooter=1290,
                engine='python'
                ).fillna(0
                        ).drop (['ages','pages', 'weight', 'packaging','size', 'pieces', 'minifigures', 'unique_pieces'], axis=1
                               ).assign(theme = lambda x: x['theme'].map(text_preprocessing_remove_LEGO))
    
df
```

```python
df_price_update = df.assign(amazon_price=lambda x: x.index.map(lambda lego_id: extract_current_amazon_price(lego_id)))
df_price_update
```

```python
df_price_update = df_price_update.drop(df_price_update[df_price_update.amazon_price == ''].index)
df_price_update['percentage'] = df_price_update.apply(lambda x: assessing_price_increase(x['price'], x['amazon_price'], x['year']), axis=1)

df_price_update
```

```python
df_price_by_theme = df_price_update.drop (['set_name', 'price', 'amazon_price', 'year'], axis=1
                                         ).groupby(['theme']).mean()
df_price_by_theme.plot (
    y=['percentage'],
    kind = 'bar',
    figsize=(15, 5),
    grid=True    
);
```
